```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})

df.head()
```

```python
df.hist(figsize=(10,6))
plt.show()
```

Insight:
The histogram shows the distribution of numerical features.
Petal features show more variation than sepal features.



```python
sns.countplot(x='species', data=df)
plt.title("Count of Each Iris Species")
plt.show()
```

Insight:
All three iris species have equal number of samples.



```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df.drop('species', axis=1))
plt.show()
```

Insight:
Box plot helps identify outliers.
Petal width and petal length show more spread.


```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.show()
```

Insight:
Petal length and petal width have strong positive correlation.
These are important features for prediction.


Important Features:
 Petal length
 Petal width
These features are most useful for classification.
